# a. Importieren der Daten in ein Pandas Dataframe

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_excel("data.xlsx")

columns = {"X1 transaction date": "transaction date",
           "X2 house age": "house age",
           "X3 distance to the nearest MRT station": "distance to the nearest MRT station",
           "X4 number of convenience stores": "number of convenience stores",
           "X5 latitude": "latitude",
           "X6 longitude": "longitude",
           "Y house price of unit area": "house price of unit area"}

df.rename(columns=columns, inplace=True)
df
```

# b. Visualisierung der Verteilung der Merkmale nach Z-Standardisierung in Boxplots

```python
plt.figure(figsize=(25, 8), dpi=150)
df_z_transformed = (df - df.mean()) / df.std()  # Dataframe mittels Z-Transformation skalieren
sns.boxplot(data=df_z_transformed)
```

# c. Visualisierung des linearen Zusammenhangs zwischen y: 'house price of unit area' und
# jedem einzelnen Merkmal xi in Streuungsdiagrammen

```python
plt.figure(figsize=(15, 8), dpi=100)
plt.title(label="Linearer Zusammenhang zwischen Merkmal x: 'transaction date' und Zielmerkmal y", fontsize=20,
          fontweight="bold")

#sns.scatterplot(x="transaction date", y="house price of unit area", data=df, color='blue')
sns.regplot(x=df['transaction date'], y=df['house price of unit area'], color='green')
```
